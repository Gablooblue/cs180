```python
#get cost
def get_path_cost(path):
    cost = 0
    for i, n in enumerate(path):
        if( i == 0 ):
            continue
        cost += adj_mat[path[i-1] - 1][path[i] - 1]
    return cost
```

```python
def hill_climb(initial_state):
    mat_len = len(initial_state)
    curr_state = initial_state.copy()
    min_state = curr_state.copy()
    min_cost = get_path_cost(min_state)
    steps = 0
   
    while True:
        prev_state = min_state.copy()
        prev_cost = min_cost
        min_state = get_min_neighbor(min_state, min_cost)
        min_cost = get_path_cost(min_state)
        steps += 1
        if prev_cost <= min_cost:
             break
    print("--------------------")
    print("State:", prev_state) 
    print("Cost:", prev_cost)
    print("Steps:", steps)
    
        
        
```

```python
def swap(state, x, y):
    state = state.copy()
    i = state[x]
    state[x] = state[y]
    state[y] = i
    return state
```

```python
def get_min_neighbor(min_state, min_cost):
    base_state = min_state.copy()
    mat_len = len(min_state)
    for i in range(mat_len):
        for j in range(mat_len):
            curr_state = swap(base_state, i, j)
            curr_cost = get_path_cost(curr_state)
            #print("Curr State:", curr_state)
            #print("Curr Cost:", curr_cost)
            if(curr_cost < min_cost):
                min_state = curr_state.copy()
                min_cost = curr_cost
        #print("Base State:", base_state)
    
    print()
    print("Min State:", min_state)
    print("Min Cost:", min_cost)
    return min_state.copy()
            
```

```python
def simulated_annealing(initial_state, t_max, t_min):
    #Shuffle State
    shuffled_state = initial_state.copy()
    random.shuffle(shuffled_state)
    curr = shuffled_state.copy()
    #for t in np.arange(t_max, t_min - 1, -(t_max/np.log(k))):
    t = t_max
    a = random.random()
    print("a:", a)
    steps = 0
    while t > t_min:
        E_c = get_path_cost(curr)
        n = get_random_neighbor(curr)
        E_n = get_path_cost(n)
        delta_E = E_c - E_n
        if delta_E > 0:
            curr = n.copy()
            print("CORRECT CHANGE")
        elif math.exp(delta_E/t) > random.random():
            curr = n.copy()
            print("RISKY CHANGE")
        steps += 1
        print("Temp:", t)
        print("State:", curr)
        print("Cost:", get_path_cost(curr))
        print()
        #t = t_max * math.pow(a, steps)
        t = t_max / math.log(steps + 2)
    print("FINAL STATE")
    print("State:", curr)
    print("Cost:", get_path_cost(curr))

    
```

```python
def get_random_neighbor(curr_state):
    neighbor_state = curr_state.copy()
    length = len(curr_state) - 1
    random_x = random.randint(0, length)
    random_y = random.randint(0, length)
    while random_x == random_y:
        random_y = random.randint(0, length)
    neighbor_state = swap(neighbor_state, random_x, random_y)
    return neighbor_state
```
